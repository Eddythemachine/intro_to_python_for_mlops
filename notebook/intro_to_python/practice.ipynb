```python
# List Methods
# Append --- Add to the end 
# Insert --- Insert at index 1

basket = []

basket.append("Goat")
basket.append("Goat Milk")
basket.append("Milf")
basket.insert(1, ["Girl", "Boy"])
print(basket)
```

```python
# Removing Items

print(f"Before Removing goat {basket}")

basket.remove("Goat")

print(f"After Removing goat {basket}")

popped = basket.pop()
print(f"After popping {basket}")
popped = basket.pop(0)
print(f"After popping 0 {basket}")
```

```python
# ❌ WRONG - Both variables point to SAME list
original = [1, 2, 3]
copy = original
copy.append(4)
print(original)  # [1, 2, 3, 4] - Oops!

# ✅ RIGHT - Create actual copy
original = [1, 2, 3]
copy = original.copy()  # or copy = original[:]
copy.append(4)
print(original)  # [1, 2, 3] - Perfect!
```

```python
# Create dictionary of squares
squares = {x: x**2 for x in range(5)}
# {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}

# Filter dictionary
numbers = {"a": 1, "b": 2, "c": 3, "d": 4}
even_squares = {k: v**2 for k, v in numbers.items() if v % 2 == 0}
# {'b': 4, 'd': 16}
```

```python
# Merge Dict 
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

merged = dict1 | dict2  # {'a': 1, 'b': 3, 'c': 4}

print(merged)
```

```python
# Instead of long if-elif chains
def process_command(command):
    actions = {
        "start": lambda: print("Starting..."),
        "stop": lambda: print("Stopping..."),
        "pause": lambda: print("Pausing...")
    }

    action = actions.get(command, lambda: print("Unknown command"))
    return action()

process_command("start")  # "Starting..."
```

```python
# Now that absolute paths are shown, we can inspect them for file metadata

import os
from os.path import abspath, join, getsize

sizes = {}

for top_dir, directories, files in os.walk('.'):
    for _file in files:
        full_path = abspath(join(top_dir, _file))
        size = getsize(full_path)
        sizes[full_path] = size
        #break

sorted_results = sorted(sizes, key=sizes.get, reverse=True)

for path in sorted_results[:10]:
    print("Path: {0}, size: {1}".format(path, sizes[path]))
```
